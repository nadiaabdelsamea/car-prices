```python
df['engine_fuel_type'].mode()
```

```python
df['engine_fuel_type']=df['engine_fuel_type'].fillna(df['engine_fuel_type'].mode()[0])
```

```python
df['engine_cylinders']=df['engine_cylinders'].fillna(df['engine_cylinders'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df=df.drop(['number_of_doors','market_category','make','model','engine_fuel_type','vehicle_style'],axis=1)
```

```python
df['engine_hp']=df['engine_hp'].fillna(df['engine_hp'].mode()[0])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df=pd.get_dummies(df, columns=['transmission_type','driven_wheels','vehicle_size'])
```

```python
df.head()
```

```python
df.tail()
```

```python
x=df.drop(['price'],axis=1)
y=df['price']
```

```python
x.head()
```

```python
y.head()
```

**linear regression**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
predict=model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,predict)
print(mse)
```

```python
root=np.sqrt(mse)
print(root)
```

```python
from sklearn.metrics import r2_score
error=r2_score(y_test,predict)
print(error)
```

```python
from sklearn.tree import DecisionTreeRegressor
model1=DecisionTreeRegressor()
```

```python
model1.fit(x_train,y_train)
```

```python
result=model1.predict(x_test)
```

```python
accuracy=r2_score(y_test,result)
print(accuracy)
```

```python
result2=model1.predict(x_train)
```

```python
accuracy2=r2_score(y_train,result2)
print(accuracy2)
```
