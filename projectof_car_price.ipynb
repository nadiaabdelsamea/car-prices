```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/mlbookcamp-code/chapter-02-car-price/data.csv")
```

```python
df.columns=df.columns.str.lower()
```

```python
df.columns=df.columns.str.replace(" ","_")
```

```python
df.head()
```

```python
df.rename(columns={"msrp":"price"},inplace=True)
```

```python
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(7,3))
sns.histplot(df.price,bins=30)
plt.show()
```

```python
df["price"].value_counts()
```

```python
import numpy as np
df['price']=np.log1p(df.price)
```

```python
plt.figure(figsize=(7,3))
sns.histplot(df.price,bins=30)
plt.show()
```

```python
df=df.drop(columns={"make","model","market_category"},axis=1)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['engine_hp'].mean()
```

```python
df['engine_hp']=df['engine_hp'].fillna(df['engine_hp'].mean())
```

```python
df['engine_cylinders'].mean()
```

```python
df['engine_cylinders']=df['engine_cylinders'].fillna(df['engine_cylinders'].mean())
```

```python
df['number_of_doors'].mean()
```

```python
df['number_of_doors']=df['number_of_doors'].fillna(df['number_of_doors'].mean())
```

```python
df.isnull().sum()
```

```python
df['engine_fuel_type'].mode()
```

```python
df['engine_fuel_type']=df['engine_fuel_type'].fillna(df['engine_fuel_type'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df['transmission_type'].value_counts()
```

```python
df['engine_fuel_type'].value_counts()
```

```python
df=pd.get_dummies(df , columns=['engine_fuel_type','transmission_type','driven_wheels','vehicle_size','vehicle_style'])
```

```python
df.head()
```

```python
x=df.drop(columns='price',axis=1)
y=df['price']
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
predict=model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error , r2_score
```

```python
error=mean_squared_error(y_test,predict)
```

```python
print(error)
print(np.sqrt(error))
```

```python
r2_score(y_test,predict)
```

```python
import joblib as jb
```

```python
jb.dump(model,"car_price_model1.pkl")
```

```python
jb.load("/content/car_price_model1.pkl")
```

```python
model2.fit(x_train,y_train)
```

```python
result2=model2.predict(x_test)
```
